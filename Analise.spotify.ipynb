```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

```python
#assumindo que a coluna 1 é o índice
df = pd.read_csv("spotify.csv",index_col=0)
df.head()
```

```python
# Visualizando as colunas do dataset

for i, col in enumerate(df.columns):
    print(f'{i} — {col}')
```

```python
# Visualização do número de linhas e colunas

n_columns = df.shape[1]
n_rows    = df.shape[0]   

print(f'Número de colunas presentes no dataset {n_columns}. \n' + 
      f'Número de linhas presentes no dataset {n_rows}.')
```

```python
#vi
print(f'Médias da colunas númericas do dataset:\n{df.mean()} ')
```

```python
#Vamos renomear a coluna "tempo" para "time", para que todas fiquem padronizadas em inglês
df.rename(columns={"tempo":"time"},inplace=True)
df.head()
```

```python
#vamos transformar a coluna length de ms para segundos
df["length"]=df["length"]/1000000
```

```python
df.head()
```

```python
# vamos descobrir a duração perfeita de um hit, fazendo a média entre o tempo de todas as músicas
mean_hit=df['length'].mean()
print(f'A duração perfeita de um hit foi de {round(mean_hit,2)} segundos')
```

```python
#os artistas com mais hits
df['artist'].value_counts()
```

```python
#vendo o df acima,vimos que rihanna tem 27 hits, quais são eles?
df[df['artist']=='Rihanna']
```

```python
#música com maior duração
maxvalue=df['length'].max()
print(f'A música com a maior duração possui {maxvalue} segundos')
df[df['length']==maxvalue].iloc[0]
```

```python
#música com mínima duração
minvalue=df['length'].min()
print(f'A música com a menor duração possui {minvalue} segundos')
df[df['length']==minvalue].iloc[0]
```

```python
#Vendo a duração média dos hits dos anos 00's, 10's e 20's
mean_2000=df[df['playlist_year']==2000]['length'].mean()
mean_2010=df[df['playlist_year']==2010]['length'].mean()
mean_2020=df[df['playlist_year']==2020]['length'].mean()
print(f'As durações médias das músicas de 2000,2010 e 2020 são, respectivamente, {round(mean_2000,2)} {round(mean_2010,2)}  {round(mean_2020,2)}')
```

```python
#Vendo o quão dançável elas são hits dos anos 00's, 10's e 20's
mean_2000=df[df['playlist_year']==2000]['danceability'].mean()
mean_2010=df[df['playlist_year']==2010]['danceability'].mean()
mean_2020=df[df['playlist_year']==2020]['danceability'].mean()
print(f'A danceabilidade de 2000,2010 e 2020 são, respectivamente, {round(mean_2000,2)} {round(mean_2010,2)}  {round(mean_2020,2)}')
```
